```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# 1. Data fiktif pangsa pasar yang sama
data_market_share = {
    'Produk': ['Produk A', 'Produk B (Kompetitor)', 'Produk C (Kita)', 'Lainnya'],
    'Pangsa Pasar (%)': [45, 28, 15, 12]
}
df_market = pd.DataFrame(data_market_share)
# Urutkan data agar bar chart lebih rapi
df_market = df_market.sort_values('Pangsa Pasar (%)', ascending=False)


# 2. Membuat HORIZONTAL BAR CHART (pilihan yang efektif)
plt.figure(figsize=(10, 6))

# Tentukan palet warna: sorot produk kita dan kompetitor
pal = []
for prod in df_market['Produk']:
    if prod == 'Produk C (Kita)':
        pal.append('orange') # Warna untuk produk kita
    elif prod == 'Produk B (Kompetitor)':
        pal.append('cornflowerblue') # Warna untuk kompetitor
    else:
        pal.append('lightgrey') # Warna untuk konteks

# Buat barplot dengan Seaborn
ax = sns.barplot(x='Pangsa Pasar (%)', y='Produk', data=df_market, palette=pal)

# Memberi judul yang bercerita & membersihkan visual
ax.set_title('Produk B Mendominasi Kompetisi, Hampir 2x Pangsa Pasar Produk Kita', fontsize=15, weight='bold', loc='left')
ax.set_xlabel('Pangsa Pasar (%)', fontsize=12)
ax.set_ylabel('') # Hapus label sumbu Y karena sudah jelas
sns.despine(left=True, bottom=True) # Hapus garis-garis sumbu yang tidak perlu

# Opsional: Tambahkan label data di ujung batang untuk presisi
for p in ax.patches:
    width = p.get_width()
    ax.text(width + 0.6, p.get_y() + p.get_height()/2., f'{width:.0f}%', va='center')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 1. Membuat data fiktif
bulan = pd.date_range('2024-01-01', periods=10, freq='MS').strftime('%B')
tiket_sebelum = [480, 510, 500, 525, 540]
tiket_sesudah = [280, 260, 255, 265, 270]
data = {
    'Bulan': bulan,
    'Jumlah Tiket': tiket_sebelum + tiket_sesudah
}
df = pd.DataFrame(data)
df['Waktu'] = pd.to_datetime(df['Bulan'] + ' 2024')

# 2. Setup Plot
sns.set_theme(style="white")
fig, ax = plt.subplots(figsize=(12, 7))

# 3. Plot Garis dan Area Utama
ax.plot(df['Waktu'], df['Jumlah Tiket'], color='#1f77b4', linewidth=2.5)
ax.fill_between(df['Waktu'], df['Jumlah Tiket'], color='#1f77b4', alpha=0.1)

# 4. Menambahkan Elemen Cerita (ini yang terpenting!)

# Garis vertikal penanda intervensi
ax.axvline(x=pd.to_datetime('2024-05-15'), color='black', linestyle='--', linewidth=1.5, ymin=0, ymax=1.0)
ax.text(pd.to_datetime('2024-05-15'), 620, 'Sistem Bantuan Otomatis\nDILUNCURKAN',
        ha='center', fontsize=11, fontweight='bold')

# Anotasi sebelum intervensi
ax.text(pd.to_datetime('2024-03-01'), 400, 'Rata-rata SEBELUM:\n~511 tiket/bulan',
        ha='center', fontsize=11, color='dimgray')

# Anotasi setelah intervensi
ax.text(pd.to_datetime('2024-08-01'), 150, 'Rata-rata SESUDAH:\n~266 tiket/bulan',
        ha='center', fontsize=11, color='#1f77b4', fontweight='bold')

# Anotasi Penurunan
penurunan = (511 - 266) / 511
ax.annotate(f'Penurunan {penurunan:.0%}',
            xy=(pd.to_datetime('2024-09-15'), 265),
            xytext=(pd.to_datetime('2024-09-15'), 400),
            arrowprops=dict(facecolor='black', arrowstyle='->', connectionstyle="arc3,rad=-0.2"),
            ha='center', fontsize=12, fontweight='bold')

# 5. Membersihkan & Memberi Judul
# Judul yang bercerita
ax.text(pd.to_datetime('2024-01-01'), 720,
        'Implementasi Sistem Bantuan Otomatis Berhasil Menekan Jumlah Tiket',
        fontsize=18, fontweight='bold', ha='left')
# Subjudul
ax.text(pd.to_datetime('2024-01-01'), 680,
        'Jumlah tiket bantuan yang diterima per bulan turun signifikan sejak peluncuran pada bulan Mei',
        fontsize=14, ha='left', color='dimgray')

# Membersihkan sumbu
sns.despine(left=True, bottom=True)
ax.set_yticks([])
ax.set_ylabel('')
ax.set_xlabel('')

# Format tanggal di sumbu X
ax.tick_params(axis='x', which='major', labelsize=11)
plt.xticks(df['Waktu'], df['Bulan'])
ax.set_ylim(0, 750)

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Estimasi dan Strukturkan Data dari Gambar
# Data ini diperkirakan dari posisi titik-titik pada grafik di buku.
data = [
    {'Kategori': 'Harga', 'Pihak': 'Perusahaan Kita', 'Indeks': 1.30},
    {'Kategori': 'Harga', 'Pihak': 'Kompetitor A', 'Indeks': 1.10},
    {'Kategori': 'Harga', 'Pihak': 'Kompetitor B', 'Indeks': 0.85},
    {'Kategori': 'Harga', 'Pihak': 'Kompetitor E', 'Indeks': 0.70},

    {'Kategori': 'Hubungan', 'Pihak': 'Perusahaan Kita', 'Indeks': 0.75},
    {'Kategori': 'Hubungan', 'Pihak': 'Kompetitor A', 'Indeks': 0.90},
    {'Kategori': 'Hubungan', 'Pihak': 'Kompetitor B', 'Indeks': 0.60},
    {'Kategori': 'Hubungan', 'Pihak': 'Kompetitor D', 'Indeks': 0.40},

    {'Kategori': 'Layanan', 'Pihak': 'Perusahaan Kita', 'Indeks': 0.95},
    {'Kategori': 'Layanan', 'Pihak': 'Kompetitor A', 'Indeks': 1.20},
    {'Kategori': 'Layanan', 'Pihak': 'Kompetitor C', 'Indeks': 0.65},

    {'Kategori': 'Kenyamanan', 'Pihak': 'Perusahaan Kita', 'Indeks': 0.80},
    {'Kategori': 'Kenyamanan', 'Pihak': 'Kompetitor A', 'Indeks': 1.15},
    {'Kategori': 'Kenyamanan', 'Pihak': 'Kompetitor D', 'Indeks': 0.55},

    {'Kategori': 'Seleksi', 'Pihak': 'Perusahaan Kita', 'Indeks': 1.05},
    {'Kategori': 'Seleksi', 'Pihak': 'Kompetitor C', 'Indeks': 1.25},
    {'Kategori': 'Seleksi', 'Pihak': 'Kompetitor D', 'Indeks': 0.60},
    {'Kategori': 'Seleksi', 'Pihak': 'Kompetitor E', 'Indeks': 1.00},
]

df = pd.DataFrame(data)

# 2. Tentukan Bentuk Marker dan Urutan Pihak
pihak_order = ['Perusahaan Kita', 'Kompetitor A', 'Kompetitor B', 'Kompetitor C', 'Kompetitor D', 'Kompetitor E']
marker_map = {
    'Perusahaan Kita': 'D', # Diamond
    'Kompetitor A': '^', # Triangle
    'Kompetitor B': 's', # Square
    'Kompetitor C': 'o', # Circle
    'Kompetitor D': 'X', # X
    'Kompetitor E': '*'  # Star (dipilih agar tidak sama dengan 'Perusahaan Kita')
}

# 3. Buat Plot
sns.set_theme(style="whitegrid")
fig, ax = plt.subplots(figsize=(10, 7))

# Gunakan stripplot dari Seaborn
plot = sns.stripplot(
    data=df,
    x='Kategori',
    y='Indeks',
    hue='Pihak',
    hue_order=pihak_order,
    s='Pihak',
    style_order=pihak_order,
    markers=marker_map,
    s=12,  # Ukuran marker
    jitter=False, # Pastikan titik-titik tidak menyebar horizontal
    ax=ax
)

# 4. Styling agar Mirip dengan Buku
# Judul dan Label
ax.set_title("Figur 3.15 Grafik Asli\nIndeks Performa Terbobot", fontsize=16, pad=20)
ax.set_xlabel('')
ax.set_ylabel('')
ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=11)


# Sumbu Y dan Grid
ax.set_ylim(0, 1.7)
ax.set_yticks([0.0, 0.5, 1.0, 1.5])
ax.grid(axis='x', which='both', b=False) # Hapus grid vertikal
ax.grid(axis='y', linestyle='-', linewidth=1) # Atur grid horizontal

# Legenda
handles, labels = ax.get_legend_handles_labels()
# Hapus judul "Pihak" dari legenda
ax.legend(handles, labels, loc='upper left', bbox_to_anchor=(-0.4, 1.0), frameon=False, fontsize=11)


plt.show()
```
