```python
import pandas as pd
import numpy as np

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from sklearn.pipeline import Pipeline
```

```python
data = load_breast_cancer(as_frame=True)
```

```python
x, y = data.data, data.target
```

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
df
```

```python
X_Train, X_Test, Y_Train, Y_Test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=42)
```

```python
# ini melakukan Scaler model
pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('knn', KNeighborsClassifier()) # placeholder yang akan diganti 
])
```

# Hyperparameter Tunning in Action
### GridSearch CV Original
CV artinya Cross Validation

```python
# kau cari di browser dan ketik ini -> knn hyperparameter grid
param_grid = {
    'knn__n_neighbors': [3, 5, 7, 9, 11],           # tanpa knn__ ini bakal error saat di fit
    'knn__weights': ['uniform', 'distance'],        # ga wajib
    'knn__metric': ['euclidean', 'manhattan']       # ga wajib
}
```

```python
grid_search = GridSearchCV(
    estimator=pipe,
    param_grid=param_grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1               #berapa banyak cpu core yg digunakan
)
grid_search.fit(X_Train,Y_Train)
```

```python
y_pred = grid_search.predict(X_Test)
print(classification_report(Y_Test, y_pred))
```

```python
print('best estimator',grid_search.best_estimator_)
print('best score', grid_search.best_score_)
```

### HyperParameter Tunning banyak GridSearch CV model sekaligus

```python
# ini melakukan Scaler model
pipe_model = Pipeline([
    ('scaler', StandardScaler()),
    ('model', KNeighborsClassifier())
])
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
# support vektor machine

param_grid_model = [
    {   # KNN
        'model': [KNeighborsClassifier()],
        'model__n_neighbors': [3, 5, 7],
        'model__weights': ['uniform', 'distance'],
        'model__p': [1, 2]
    },
    {   # SVM
        'model': [SVC()],
        'model__C': [0.1, 1, 10],
        'model__kernel': ['linear', 'rbf'],
        'model__gamma': ['scale', 'auto']
    },
    {   # Random Forest
        'model': [RandomForestClassifier(random_state=42)],
        'model__n_estimators': [50, 100],
        'model__max_depth': [None, 10, 20]
    },
    {   # Logistic Regression
        'model': [LogisticRegression(max_iter=10000, solver='liblinear')],
        'model__C': [0.01, 0.1, 1, 10],
        'model__penalty': ['l1', 'l2']
    }
]
```

```python
grid_search_model = GridSearchCV(
    estimator=pipe_model,
    param_grid=param_grid_model,
    cv=2,
    scoring='accuracy',
    n_jobs=-1  
)
grid_search_model.fit(X_Train,Y_Train)
```

```python
print('Best Model & Params:\n')
print(grid_search_model.best_params_['model'],'\n')
print(grid_search_model.best_params_)
print(f'Best CV Accuracy" {grid_search_model.best_score_:.4f}')
```

```python
y_pred = grid_search_model.predict(X_Test)
print(classification_report(Y_Test, y_pred))
```

# Random Search CV

```python
pipe_random = Pipeline([
    ('scaler', StandardScaler()),
    ('knn', KNeighborsClassifier())
])
```

```python
param_grid_random = {
    'knn__n_neighbors': [3,5,7,9,11]
}
```

```python
random_seacrh = RandomizedSearchCV(
    estimator=pipe_random,
    param_distributions=param_grid_random,
    n_iter=3,
    random_state=42,
    cv=5,
    scoring='accuracy',
    n_jobs=-1
)
random_seacrh.fit(X_Train,Y_Train)
```

```python
y_predRandom = random_seacrh.predict(X_Test)
print(classification_report(Y_Test, y_predRandom))
```

```python
print(random_seacrh.best_estimator_,'\n')
print(random_seacrh.best_score_)
```
