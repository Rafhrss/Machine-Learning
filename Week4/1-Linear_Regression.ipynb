```python
import pandas as pd
import numpy as np

from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data = fetch_california_housing(as_frame=True)
```

```python
df = data.frame
```

```python
df.sample(10)
```

```python
x = df.drop('MedHouseVal', axis=1)
y = df['MedHouseVal']
```

```python
X_Train, X_test, Y_Train, Y_Test = train_test_split(x, y , test_size=0.2, random_state=42)
# kenapa tidak ada stratify=y karena tidak ada yg perlu di seimbangkan.
# karena gada sebuah kategori yg harus di balance kan
```

## Linear Regression

```python
model = LinearRegression()
```

```python
model.fit(X_Train, Y_Train)
```

```python
y_pred_lin = model.predict(X_test )
```

```python
print(y_pred_lin[0])
```

```python
print(Y_Test[0])
```

kenapa nilai prediksi nya 0.719122841 sedangkan nilai Test nya 4.5<br>
apakan ini nilai yg bagus atau tidak, dan ini alasanya. peluang paling tinggi adalah 1<br>
gunakan min() dan max() untuk melihat nya. btw aga kurang paham jg ak<br>


```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

```python
mae = mean_absolute_error(Y_Test, y_pred_lin)
```

```python
print(mae)
# 0.5 dibilang kecil karena ---- liat aja min dan max nya prediksi dibawah ini
```

### Norm Mean Abs Error -> Normalisasi -> Min Max Scaling = Persentase.Min = 0, Max = 1

```python
y_pred_lin.max()
```

```python
y_pred_lin.min()
```

```python
range = y_pred_lin.max() - y_pred_lin.min()
NMAE = mae / range
```

```python
print(NMAE)
```
